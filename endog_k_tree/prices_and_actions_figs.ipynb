```python
import numpy as np
import matplotlib.pyplot as plt

from endog_k_price import RP
from graph_builder import *
```

```python
ps = RP(n=200, beta=0.005, delta=1.025, c=lambda x: np.exp(5 * x) - 1)
ps.set_prices()

fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(ps.grid, ps.p)

plt.show()
```
