```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
from endog_k_price import RP
from graph_builder import *
```

```python
from networkx.drawing.nx_agraph import graphviz_layout
```

Build the graph. 

```python
def draw_graph(ps, scale_factor=2000, figsizes=(15, 15)):
    
    firms, G = build_dict_and_graph(ps)
    pos = graphviz_layout(G, prog='twopi') # twopi or dot

    fig, ax = plt.subplots(figsize=figsizes)
    ax.axis('off', frame_on=False)

    vas = []
    for firm in firms.values():
        vas.append(firm.val_add * scale_factor)

    nx.draw(G, 
            pos, 
            ax=ax,
            node_size=vas, 
            node_shape='o',
            alpha=0.4, 
            node_color="blue", 
            with_labels=False)

    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
```

```python
ps = RP(n=250, beta=0.005, delta=1.025, c=lambda x: np.exp(5 * x) - 1)
ps.set_prices()
draw_graph(ps)
```

```python
ps = RP(n=250, beta=0.005, delta=1.01, c=lambda x: np.exp(5 * x) - 1)
ps.set_prices()
draw_graph(ps)
```

```python
ps = RP(n=200, beta=0.0001, delta=1.02, c=lambda x: x**1.1)
ps.set_prices()
draw_graph(ps, figsizes=(15,15), scale_factor=100000)
```
